1. Start with beta=1, sigma=1. Simulate one at a time to see how the histogram gradually builds up.
2. With beta=1, sigma=1, simulate 5000 experiments at once
3. With beta=1, sigma=0.2, simulate 5000 experiments at once
4. With beta=1, sigma=3, simualte 5000 experiments at once

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
rng = np.random.default_rng(123)
```

```python
estimates = []
beta = 1
sigma = 1
N = 100   # number of data points
```

```python
for i in range(1000):

    X = rng.uniform(-1,1,size=(N,1))
    eps = rng.normal(0,sigma,size=(N,1))
    Y = beta*X + eps
    beta_hat = np.linalg.inv(np.matmul( X.transpose(), X)) * np.matmul( X.transpose(), Y)
    estimates.append(
        beta_hat[0,0]
    )

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(X,Y)
plt.plot(X, beta_hat*X,color='red')
plt.xlim([-1,1])
plt.ylim([-5,5])

plt.subplot(1,2,2)
plt.hist(estimates)
plt.xlim([0,2])

plt.show()
```
