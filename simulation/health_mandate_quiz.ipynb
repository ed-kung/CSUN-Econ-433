Code for generating the graph for health mandate quiz

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
rng = np.random.default_rng(100)

df = []
for year in range(2002, 2012):
    if year==2002:
        y0 = 0
        y1 = 0
    else:
        y0 = rng.uniform(-0.03,0.03)
        y1 = (year-2002)*0.025 + rng.uniform(-0.03,0.03)
    df.append({'year':year, 'y0':y0, 'y1':y1})

df = pd.DataFrame.from_dict(df)
```

```python
plt.plot(df['year'], df['y0'], marker='.')
plt.plot(df['year'], df['y1'], linestyle='dashed', marker='.')
plt.axvline(x=2006, color='red', alpha=0.5)
plt.axvline(x=2008, color='red', alpha=0.5)
plt.legend(['Non-MA', 'MA'])
plt.grid(alpha=0.2)
plt.xlabel('Year')
plt.ylabel('Health insurance coverage in population')
plt.show()
```
